```python
from bs4 import BeautifulSoup as soup
import requests
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url = 'http://www.worldstopexports.com/wine-exports-country/'
```

```python
response = requests.get(url)
soup = soup(response.content, "html.parser")
```

```python
num_int = len(soup.find_all("td", class_ = "column-2"))

exporter = []

for i in range(num_int):
    countries = soup.find_all("td", class_ = "column-2")[i].get_text()
    exporter.append(countries)
    
exporter    
```

```python
num_int = len(soup.find_all("td", class_ = "column-3"))

exported_wine = []

for i in range(num_int):
    q = soup.find_all("td", class_ = "column-3")[i].get_text()
    exported_wine.append(q)
```

```python
#deleting '$' from exported_wine column
exported_wine2 = []

for i in exported_wine:
    exported_wine2.append(i[1:])

exported_wine2
```

```python
#deleting ',' from wine_exported column
exported_wine3 = [i.replace(",", "") for i in exported_wine2]
exported_wine3
```

```python
#list of int
exported_wine4 = [int(i)/1000 for i in exported_wine3]
exported_wine4
```

```python
num_int = len(soup.find_all("td", class_ = "column-4"))

variation = []

for i in range(num_int):
    z = soup.find_all("td", class_ = "column-4")[i].get_text()
    variation.append(z)
    
variation 
```

```python
wine_exports = pd.DataFrame(list(zip(exporter, exported_wine4)), 
               columns = ['Country', 'Exported_wine(US$)']) 
wine_exports
```

```python
#country column in wines df = country column in web_scraping
wine_exports.Country.replace({'United States':'US', 'England':'United Kingdom', 'Bosnia and Herzegovina':'Bosnia/Herzegovina'}, inplace = True)
wine_exports
```

```python
wine_exports.to_csv('../data/wines_webscraping.csv')
```
